ROC curve

```python
outer_cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=SEED)

y_true_all = []
y_proba_all = []

for train_idx, test_idx in outer_cv.split(X, y):
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

    # Fit GridSearchCV on the inner folds using pipeline
    clf_gs.fit(X_train, y_train)

    # Best pipeline from this fold
    best_model = clf_gs.best_estimator_

    # Predict probabilities on the test split
    y_proba = best_model.predict_proba(X_test)[:, 1]

    # Store true labels and predicted probs
    y_true_all.extend(y_test)
    y_proba_all.extend(y_proba)

# Compute ROC curve
fpr, tpr, thresholds = roc_curve(y_true_all, y_proba_all)
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {roc_auc:.2f})', color='darkorange')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - Nested CV with Random Forest')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```
